### Wordnet Lemmatizer
It is like stemming, the o/p will be 'lemma' which is a root word rather than the root stem, the o/p of stemming.After lemmatization we will get a valid word that means the same thing.
NLTK provides WordNetLemmatizer classwhich is a thin wrapper around the wordnet corpus.This class uses morphy() functionto the WOrdNet CorpusReader class to find a lemma.

Uses:
Q&A, chatbots, text summarization

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going", pos="v")
```

```python
'''
POS- Noun n
verb- v
adjective- a
adverb-r
'''
```

```python
words = ["eating","eats","eaten","writing","writes","programming","program","history","finally","finalized"]
```

```python
for word in words:
    print(word+"--->>>"+lemmatizer.lemmatize(word, pos='v'))
```
