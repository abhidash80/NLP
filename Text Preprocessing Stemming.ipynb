### Stemming
Process of reducing a word to its word stem that affixes to suffuixes and prefixes or to the roots of words known as lemma.

```python
#Classification Problem
##Comments of product is a positive review or negative review
##Review  ---->>> eating, eat, eaten [going, gone, goes]   ---->> go
words = ["eating","eats","eaten","writing","writes","programming","program","history","finally","finalize"]
```

### Stemming techniques
PorterStemmer

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"--->>>"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem("sitting")
```

### RegexpStemmer class
Regular expression Stemmer algorithms-->> Takes single regx and removesany prefix or suffix

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word+"--->>>"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
snowballstemmer.stem("fairly"),snowballstemmer.stem("sportingly")
```

```python
snowballstemmer.stem('goes')
```
