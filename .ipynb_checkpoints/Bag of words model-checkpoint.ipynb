```python
##Data Cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(msg)):
    review=re.sub('[^a-zA-Z]',' ',msg['message'][i])
    review = review.lower()
    review=review.split()
    review=[ps.stem(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
##Create Bag of words model using CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

##for binary BoW enable binary flag
cv=CountVectorizer(max_features=2500, binary=True)   ##take top 2500 vocabularies
```

```python
X=cv.fit_transform(corpus).toarray()
```

### N-grams

```python
### Create Bag of words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BoW enable binary=True
cv = CountVectorizer(max_features=1000, binary=True, ngram_range=(1,2))
Y = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(max_features=1000, binary=True, ngram_range=(2,2))
Y = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(max_features=1000, binary=True, ngram_range=(3,3))  ### more accuracy
Y = cv.fit_transform(corpus).toarray()
```
